```python
%sql

-- 1. Should show 6 tables
SHOW TABLES IN bfsi_lakehouse.metadata;
```

```python
%sql

-- 2. Inspect CHECK constraints on table_config
DESCRIBE EXTENDED bfsi_lakehouse.metadata.table_config;
SHOW TBLPROPERTIES bfsi_lakehouse.metadata.table_config;

DESCRIBE EXTENDED bfsi_lakehouse.metadata.table_process_config;
DESCRIBE EXTENDED bfsi_lakehouse.metadata.input_column_config;
```

```python
%sql

-- 3. All CHECK constraints across the metadata schema
SELECT table_name, constraint_name, constraint_type
FROM bfsi_lakehouse.information_schema.table_constraints
WHERE table_schema = 'metadata'
ORDER BY table_name, constraint_type;

```

```python
%sql
-- 4. FK relationships
SELECT
    tc.table_name AS child_table,
    kcu.column_name AS child_column,
    ccu.table_name AS parent_table,
    ccu.column_name AS parent_column
FROM bfsi_lakehouse.information_schema.referential_constraints rc
JOIN bfsi_lakehouse.information_schema.table_constraints tc
    ON rc.constraint_name = tc.constraint_name
JOIN bfsi_lakehouse.information_schema.key_column_usage kcu
    ON tc.constraint_name = kcu.constraint_name
JOIN bfsi_lakehouse.information_schema.constraint_column_usage ccu
    ON rc.unique_constraint_name = ccu.constraint_name
WHERE tc.table_schema = 'metadata';
```
