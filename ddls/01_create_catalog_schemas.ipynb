```python
%sql
-- ================================================================
-- BFSI Lakehouse — Catalog & Schema Bootstrap
-- ================================================================

-- Catalog
CREATE CATALOG IF NOT EXISTS bfsi_lakehouse
  COMMENT 'BFSI Lakehouse — synthetic NBFC dataset for DE practice';

-- Schemas (one per layer + metadata)
CREATE SCHEMA IF NOT EXISTS bfsi_lakehouse.metadata
  COMMENT 'Pipeline configuration and run logs';
CREATE SCHEMA IF NOT EXISTS bfsi_lakehouse.raw
  COMMENT 'Synthetic Parquet generator output (landing zone)';
CREATE SCHEMA IF NOT EXISTS bfsi_lakehouse.bronze
  COMMENT 'As-is Delta copies of raw — schema-enforced, idempotent';
CREATE SCHEMA IF NOT EXISTS bfsi_lakehouse.silver
  COMMENT 'Cleaned, deduplicated, parsed Delta tables';
CREATE SCHEMA IF NOT EXISTS bfsi_lakehouse.gold
  COMMENT 'Business-ready aggregates and marts';

-- Volume for raw landing zone (synthetic generator writes here)
DROP VOLUME IF EXISTS bfsi_lakehouse.raw.synthetic_data;
CREATE VOLUME IF NOT EXISTS bfsi_lakehouse.raw.synthetic_data
  COMMENT 'Synthetic Parquet, partitioned by dt + batch';

-- Volumn for Project Documents
CREATE VOLUME IF NOT EXISTS bfsi_lakehouse.default.docs
  COMMENT 'Architecture diagrams and project documentation';

-- Verify
SHOW SCHEMAS IN bfsi_lakehouse;
```
